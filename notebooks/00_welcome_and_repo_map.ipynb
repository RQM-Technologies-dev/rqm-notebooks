## What is the RQM ecosystem?

The RQM stack is a layered approach to quantum computing that grounds everything in the mathematics of quaternions and SU(2) symmetry:

```
rqm-core  ──►  rqm-qiskit  ──►  rqm-notebooks
```

| Package | Role |
|---|---|
| **rqm-core** | Canonical math: quaternions, spinors, Bloch vectors, SU(2) |
| **rqm-qiskit** | Execution bridge: rqm-core ↔ Qiskit circuits |
| **rqm-notebooks** | This repo — polished demos and learning materials |
