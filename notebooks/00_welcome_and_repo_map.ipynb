## What is this repo?

`rqm-notebooks` is the **interactive demonstration layer** of the RQM ecosystem.

It is not a standalone library — it is a structured collection of Jupyter notebooks that make the mathematics and execution machinery of the RQM stack visible, teachable, and directly usable.

If you want to *understand* quaternion geometry and how it connects to quantum computing, you are in the right place.


## The RQM stack

The RQM ecosystem is a layered approach to quantum computing grounded in quaternion mathematics and SU(2) symmetry:

```
rqm-core  →  rqm-qiskit  →  rqm-notebooks
maths        execution       demos / learning
```

| Package | Role |
|---|---|
| **rqm-core** | Canonical math: quaternions, spinors, Bloch vectors, SU(2) operations |
| **rqm-qiskit** | Execution bridge: rqm-core ↔ Qiskit circuits and simulators |
| **rqm-notebooks** | This repo — polished demos, visualisations, and learning materials |

### Why are the repos separated?

Separation keeps each concern clean:

- `rqm-core` is pure mathematics. It has no Qiskit dependency and no display logic.
- `rqm-qiskit` translates math objects into runnable circuits. It has no display logic.
- `rqm-notebooks` imports from both but adds *only* display helpers. It never redefines canonical math.

This means you can use `rqm-core` in any Python project without pulling in Jupyter, and you can add new notebooks without touching the math.

> **Key principle:** Notebooks import the math; they do not redefine it.


## Simulator-first approach

All notebooks in this series run on a **local Qiskit Aer simulator**.

This means:
- No IBM account required
- No internet connection required to run notebooks
- Results are deterministic (given a fixed seed) or statistically expected
- You can iterate quickly without queuing on real hardware

The IBM-ready path is introduced in `09_ibm_ready_path.ipynb`. It shows how to adapt the same circuits for real hardware — but everything before that works entirely offline.
